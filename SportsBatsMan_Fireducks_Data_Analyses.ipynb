```python
import fireducks.pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
```

```python

df = pd.read_csv('12. Seaborn_Batsman (1).csv')
print(df.head())
```

```python
df.info()
```

```python
df.describe(include="all")
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'runs' is the column containing the runs scored
plt.figure(figsize=(10, 6))
sns.histplot(df['RunsScored'], kde=True)  # Use 'runs' column, kde for density curve
plt.title('Frequency Distribution of Runs Scored')
plt.xlabel('Runs')
plt.ylabel('Frequency')
plt.show()
```

```python
# Assuming 'BattingPosition' and 'RunsScored' are the correct column names
plt.figure(figsize=(10, 6))
sns.boxplot(x='BattingPosition', y='RunsScored', data=df)
plt.title('Batting Position vs. Runs Scored')
plt.xlabel('Batting Position')
plt.ylabel('Runs Scored')
plt.show()

# Calculate the correlation
correlation = df['BattingPosition'].corr(df['RunsScored'])
print(f"Correlation between Batting Position and Runs Scored: {correlation}")
```

```python
# Group by batting position and calculate the mean runs scored
position_runs = df.groupby('BattingPosition')['RunsScored'].mean()

# Find the position with the highest mean runs
best_position = position_runs.idxmax()
max_runs = position_runs.max()

print(f"The best batting position for maximizing runs scored is position {best_position}, with an average of {max_runs} runs.")

# Visualization
plt.figure(figsize=(10, 6))
sns.barplot(x=position_runs.index, y=position_runs.values)
plt.title('Average Runs Scored per Batting Position')
plt.xlabel('Batting Position')
plt.ylabel('Average Runs Scored')
plt.show()
```

```python

import matplotlib.pyplot as plt
import seaborn as sns


# 1. Runs Scored Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['RunsScored'], kde=True)
plt.title('Distribution of Runs Scored')
plt.xlabel('Runs Scored')
plt.ylabel('Frequency')
plt.show()

# 2. Runs Scored vs. Balls Faced (Scatter Plot)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='BallsFaced', y='RunsScored', data=df)
plt.title('Runs Scored vs. Balls Faced')
plt.xlabel('Balls Faced')
plt.ylabel('Runs Scored')
plt.show()


# 3. Batting Position vs. Runs Scored (Box Plot)
plt.figure(figsize=(10, 6))
sns.boxplot(x='BattingPosition', y='RunsScored', data=df)
plt.title('Batting Position vs. Runs Scored')
plt.xlabel('Batting Position')
plt.ylabel('Runs Scored')
plt.show()

# 4. Dismissal Type Distribution (Count Plot)
plt.figure(figsize=(10, 6))
sns.countplot(x='Dismissal', data=df)
plt.title('Distribution of Dismissal Types')
plt.xlabel('Dismissal Type')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels if needed
plt.show()

# 5. Runs Scored by Inning (Box Plot or Violin Plot)
plt.figure(figsize=(10, 6))
sns.boxplot(x='Inning', y='RunsScored', data=df) # or sns.violinplot
plt.title('Runs Scored by Inning')
plt.xlabel('Inning')
plt.ylabel('Runs Scored')
plt.show()


# 6. Performance against Opposition (Box Plot or Violin Plot)
plt.figure(figsize=(12, 6))
sns.boxplot(x='Opposition', y='RunsScored', data=df) # or sns.violinplot
plt.title('Performance Against Different Oppositions')
plt.xlabel('Opposition')
plt.ylabel('Runs Scored')
plt.xticks(rotation=45, ha='right')
plt.show()


# 7. Runs Scored Trend Over Years (Line Plot)
runs_per_year = df.groupby('Year')['RunsScored'].sum()
plt.figure(figsize=(10, 6))
sns.lineplot(x=runs_per_year.index, y=runs_per_year.values)
plt.title('Total Runs Scored Over Years')
plt.xlabel('Year')
plt.ylabel('Total Runs Scored')
plt.show()


# 8. Heatmap for correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(df[['RunsScored','BallsFaced','BattingPosition','Inning','Year']].corr(), annot=True, cmap='coolwarm')
plt.show()
```
